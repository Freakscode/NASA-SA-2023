```python
from PIL import Image
from pydub import AudioSegment
from pydub.generators import Sine

# Definir una escala de notas para Do mayor
c_major_scale = [130.81, 146.83, 164.81, 174.61, 196.00, 220.00, 246.94, 261.63]  # Do, Re, Mi, Fa, Sol, La, Si, Do

def image_to_sound(image_path, block_size=10):
    # Cargar la imagen
    img = Image.open(image_path).convert('L')  # Convertir a escala de grises
    img_data = img.load()

    # Crear un segmento de audio vacío
    audio = AudioSegment.empty()

    # Recorrer cada bloque de píxeles en la imagen
    for y in range(0, img.height - block_size, block_size):
        for x in range(0, img.width - block_size, block_size):
            # Calcular el valor medio del bloque de píxeles
            block_value = 0
            for i in range(block_size):
                for j in range(block_size):
                    block_value += img_data[x+i, y+j]
            block_value /= (block_size * block_size)

            # Mapear el valor del bloque a una nota en la escala musical
            note = c_major_scale[int(block_value / (256 / len(c_major_scale)))]

            # Generar un tono con esa nota y añadirlo al segmento de audio
            tone = Sine(note).to_audio_segment(duration=10)  # Duración en milisegundos
            audio += tone

    # Guardar el audio en un archivo .wav
    audio.export("output.wav", format="wav")
```

```python
from moviepy.editor import VideoFileClip
from PIL import Image
import numpy as np
from pydub import AudioSegment
from pydub.generators import Sine

# Definir una escala de notas para Do mayor
c_major_scale = [130.81, 146.83, 164.81, 174.61, 196.00, 220.00, 246.94, 261.63]  # Do, Re, Mi, Fa, Sol, La, Si, Do

def video_to_sound(video_path):
    # Cargar el vídeo
    video = VideoFileClip(video_path)

    # Crear un segmento de audio vacío
    audio = AudioSegment.empty()

    # Recorrer cada fotograma del vídeo
    for frame in video.iter_frames():
        # Convertir el fotograma a una imagen PIL y convertirlo a escala de grises
        img = Image.fromarray(frame).convert('L')
        img_data = np.array(img)

        # Dividir la imagen en segmentos y procesar cada segmento
        segments = np.array_split(img_data, img_data.shape[0] // 50)
        for segment in segments:
            # Mapear el valor medio del segmento a una nota en la escala musical
            note = c_major_scale[int(np.mean(segment) / (256 / len(c_major_scale)))]

            # Generar un tono con esa nota y añadirlo al segmento de audio
            tone = Sine(note).to_audio_segment(duration=10)
            audio += tone

    # Guardar el audio en un archivo .wav
    audio.export("output.wav", format="wav")
```
